```python
# Import Libraries and Dependencies
import pandas as pd
from datetime import datetime
from pathlib import Path
data1 = Path("./Resources/athletic_sales_2020.csv")
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
data1_read = pd.read_csv(data1)
data1_read
```

```python
data2 = Path("./Resources/athletic_sales_2021.csv")
data2_read = pd.read_csv(data2)
```

```python
# Display the 2020 sales DataFrame
data2_read
```

```python
# Display the 2021 sales DataFrame
data1_read
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
data2_read.info()
```

```python
# Check the 2021 sales data types.
data1_read.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
data_combine = pd.concat([data1_read, data2_read]).reset_index(drop=True)
```

```python
# Check if any values are null.
data_combine.isnull().sum()
```

```python
# Check the data type of each column
data_combine.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
data_combine['invoice_date'] = pd.to_datetime(data_combine['invoice_date'], infer_datetime_format=True)
```

```python
# Confirm that the "invoice_date" data type has been changed.
data_combine.dtypes
```
